# Liner Regression




This code generates a synthetic regression dataset with two features, visualizes the relationship between each feature and the target variable $y$, fits a linear regression model, and attempts to plot regression lines for each feature. Let's break it down step by step:

---

**Step-by-Step Explanation** **1. Data Generation**

```python
X, y, coefs = make_regression(n_samples=100, n_features=2, noise=10, bias=10, coef=True)
```

- `X`: A 2D array of shape `(100, 2)` with two features.

- `y`: A 1D array of target values with added Gaussian noise (`noise=10`) and a bias term (`bias=10`).

- `coefs`: The true coefficients used to generate the target variable $y$


---

**2. Scatter Plot for Feature 1**

```python
plt.scatter(X[:, 0], y, color="blue")
plt.show()
```

- Visualizes the relationship between the first feature (`X[:, 0]`) and the target variable $y$.


---

**3. Scatter Plot for Feature 2**

```python
plt.scatter(X[:, 1], y, color="blue")
plt.show()
```

- Visualizes the relationship between the second feature (`X[:, 1]`) and $y$.


---

**4. Fitting a Linear Regression Model**

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
```

- Creates a `LinearRegression` model.

- Fits the model to the entire dataset $X$ (both features) and $y$.
**Key Outputs** :
- `reg.coef_`: Contains the coefficients of the regression model, one for each feature.

- `reg.intercept_`: The y-intercept of the regression model.


---

**5. Plotting Regression Line for Feature 1**

```python
x_line = np.linspace(X.min(), X.max(), 100)  # Generate x values
y0_line = reg.coef_[0] * x_line + reg.intercept_  # Line equation
plt.scatter(X[:, 0], y, color='blue', alpha=0.6, label='Data Points')  # Scatter plot for Feature 1
plt.plot(x_line, y0_line, color='red', linewidth=2, label='Regression Line')  # Regression line
plt.show()
```

- Generates a range of $x$-values for plotting the regression line.

- Computes the corresponding $y$-values for feature 1 using:
y = \text{reg.coef_[0]} \cdot x + \text{reg.intercept_}

- Visualizes the data points for feature 1 and plots the regression line.


---

**6. Plotting Regression Line for Feature 2**

```python
x_line = np.linspace(X.min(), X.max(), 100)  # Generate x values
y1_line = reg.coef_[1] * x_line + reg.intercept_  # Line equation
plt.scatter(X[:, 1], y, color='blue', alpha=0.6, label='Data Points')  # Scatter plot for Feature 2
plt.plot(x_line, y1_line, color='red', linewidth=2, label='Regression Line')  # Regression line
plt.show()
```

- Similar to feature 1, but for feature 2.

- Uses the coefficient for feature 2 (`reg.coef_[1]`) and the intercept (`reg.intercept_`).

````python
from sklearn.datasets import make_regression
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
````

````python
X, y, coefs = make_regression(n_samples=100, n_features=2, noise=10, bias=10, coef=True)
````

````python
plt.scatter(X[:, 0], y, color="blue")
plt.show()
plt.scatter(X[:, 1], y, color="blue")
plt.show()
````

````python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(X, y)
````

````python
# Generate x-values for the line
x_line = np.linspace(X.min(), X.max(), 100)  # 100 points between min and max of X
y0_line = reg.coef_[0] * x_line + reg.intercept_  # Line equation
plt.scatter(X[:, 0], y, color='blue', alpha=0.6, label='Data Points')
plt.plot(x_line, y0_line, color='red', linewidth=2, label='Regression Line')
plt.show()

x_line = np.linspace(X.min(), X.max(), 100)  # 100 points between min and max of X
y1_line = reg.coef_[1] * x_line + reg.intercept_  # Line equation
plt.scatter(X[:, 1], y, color='blue', alpha=0.6, label='Data Points')
plt.plot(x_line, y1_line, color='red', linewidth=2, label='Regression Line')
plt.show()
````

# Polynomial Regression



The `PolynomialFeatures` class in `sklearn.preprocessing` generates a feature matrix by including all polynomial combinations of the original features up to the specified degree.For **degree=2** , the `fit_transform` method generates a matrix with the following features:

**Example:**

*Input:*

If $X$ is the input matrix with shape $(n, m)$, where:
- $n$: Number of samples (rows).

- $m$: Number of features (columns).

*Output:*

The resulting matrix $X_{\text{poly}}$ will have the shape $(n, k)$, where $k$ is calculated as:$
 k = \frac{(m + d)!}{m! \cdot d!}
$
Here, $d$ is the degree of the polynomial.For **degree=2**  and a single feature ($m = 1$), the transformation includes:
1. The original feature ($x_1$).

2. The square of the feature ($x_1^2$).

3. The bias term (constant 1) if `include_bias=True`.


**Example:**

*Input:*

If $X = \begin{bmatrix} 2 \\ 3 \end{bmatrix}$, then `fit_transform` with degree=2 generates:

*Output:*

$
 X_{\text{poly}} = \begin{bmatrix} 1 & 2 & 4 \\ 1 & 3 & 9 \end{bmatrix}
$


**Explanation:**

Each row corresponds to:
$
 \text{[1, } x_1, \, x_1^2 \text{]}
$

For:

- $x_1 = 2$, it becomes $[1, 2, 4]$.

- $x_1 = 3$, it becomes $[1, 3, 9]$.



**Multidimensional Input:**

For a 2-feature input:
$
 X = \begin{bmatrix} x_1 & x_2 \\ \end{bmatrix}
$
`fit_transform` generates terms for all combinations:$
 X_{\text{poly}} = [1, x_1, x_2, x_1^2, x_1x_2, x_2^2]
$

**Example:**

For $X = \begin{bmatrix} 2 & 3 \\ \end{bmatrix}$, it becomes:$
 X_{\text{poly}} = \begin{bmatrix} 1 & 2 & 3 & 4 & 6 & 9 \end{bmatrix}
$






````python
from sklearn.datasets import make_regression
from sklearn.preprocessing import PolynomialFeatures
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np

X, y = make_regression(n_samples=100, n_features=1, noise=10, bias=10)
````

````python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

plt.scatter(X, y, color="blue")
plt.show()

print(X_poly.shape)

plt.scatter(X_poly[:, 2], y, color="green")
plt.show()
````

````python
model = LinearRegression()
model.fit(X_poly, y)

print("Intercept:", model.intercept_)  # Beta_0
print("Coefficients:", model.coef_)   # Beta_1, Beta_2
````


- **Coefficients** : 
  - First coefficient ($0$): Corresponds to the bias term $1$. Since its already accounted for by the intercept, this coefficient is always $0$ and doesn't contribute directly to the equation.
 
  - Second coefficient ($92.41$): Corresponds to $X$ (the linear term, $\beta_1$).
 
  - Third coefficient ($0.358$): Corresponds to $X^2$ (the quadratic term, $\beta_2$).
 
$
 y = 9.023 + 92.41 \cdot X + 0.358 \cdot X^2 
$
