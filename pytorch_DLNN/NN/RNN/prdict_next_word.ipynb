```python
import torch
import datasets
import pathlib
import tokenizers
from tokenizers.trainers import BpeTrainer
from tokenizers.models import BPE
from tokenizers.normalizers import Sequence, NFKC, Lowercase
from tokenizers.pre_tokenizers import Whitespace
```

```python
import test


text_data_path = pathlib.Path("/home/xd/Documents/py_codes/ML_Notebooks/dataset/text_data")
all_data_files = [str(file) for file in text_data_path.glob("*.txt")]

split = 0.8

dataset = datasets.load_dataset("text", data_files=all_data_files)
train_dataset, test_dataset = dataset["train"].train_test_split(test_size=1 - split).values()

len(train_dataset), len(test_dataset)

print(train_dataset.info)
print(test_dataset.info)
```

```python
def batch_iterator(dataset, batch_size=1000):
  for i in range(0, len(dataset), batch_size):
    yield dataset[i : i+batch_size]["text"]
```

```python
tokenizer = tokenizers.Tokenizer(BPE(unk_token="[UNK]"))

tokenizer.normalizer = Sequence([NFKC(), Lowercase()])
tokenizer.pre_tokenizer = Whitespace()

trainier = BpeTrainer(vocab_size=30000,
                      min_frequency=2,
                      special_tokens=["[PAD]", "[UNK]", "[CLS]", "[SEP]", "[MASK]"])

tokenizer.train_from_iterator(iterator=batch_iterator(dataset=train_dataset, batch_size=10000), trainer=trainier)

tokenizer.save("BPE_tokenizer_v2.json")
```
